```markdown
## Výběr Postavy: Kdo dnes večer budeš?

Než se pustíš do pátečního dobrodružství, vyber si svého avatara. Každá postava má své silné a slabé stránky, které ovlivní tvé počáteční statistiky a možná i průběh večera.

1.  **Pracoholik Petr:** Týden ho vyždímal, ale mozek mu stále šlape na plné obrátky. Je unavený, ale racionální.
    *   **Energie:** 40
    *   **Nálada:** 60
    *   **Opilost:** 0
    *   **Příčetnost:** 90
    *   **Hlad:** 50

2.  **Párty Pavla:** Už ve čtvrtek plánovala, co se bude dít. Je plná energie, ale občas ne úplně rozumná.
    *   **Energie:** 80
    *   **Nálada:** 90
    *   **Opilost:** 0
    *   **Příčetnost:** 60
    *   **Hlad:** 30

3.  **Introvert Ivan:** Těšil se na klidný večer doma. Nechce se mu mezi lidi, ale je v pohodě.
    *   **Energie:** 70
    *   **Nálada:** 70
    *   **Opilost:** 0
    *   **Příčetnost:** 80
    *   **Hlad:** 40

4.  **Gurmet Gábina:** Celý týden snila o dobrém jídle. Je hladová, ale pozitivní.
    *   **Energie:** 60
    *   **Nálada:** 80
    *   **Opilost:** 0
    *   **Příčetnost:** 75
    *   **Hlad:** 90

Vyber si svou postavu (1-4):
```

````python
import random

def generate_random_consequence(consequences_list):
    """
    Vybere náhodný následek ze seznamu.
    """
    return random.choice(consequences_list)


def display_stats(player):
    print("\n--- Aktuální Stav Hrdiny ---")
    print(f"Energie: {player['energie']}")
    print(f"Nálada: {player['nalada']}")
    print(f"Opilost: {player['opilost']}")
    print(f"Příčetnost: {player['pricitnost']}")
    print(f"Hlad: {player['hlad']}")
    print("---------------------------\n")


player = {
    'jmeno': '',
    'energie': 0,
    'nalada': 0,
    'opilost': 0,
    'pricitnost': 0,
    'hlad': 0
}

volba_postavy = input("Zadejte číslo vaší volby (1-4): ")

if volba_postavy == '1':
    player['jmeno'] = 'Pracoholik Petr'
    player['energie'] = 40
    player['nalada'] = 60
    player['opilost'] = 0
    player['pricitnost'] = 90
    player['hlad'] = 50
elif volba_postavy == '2':
    player['jmeno'] = 'Párty Pavla'
    player['energie'] = 80
    player['nalada'] = 90
    player['opilost'] = 0
    player['pricitnost'] = 60
    player['hlad'] = 30
elif volba_postavy == '3':
    player['jmeno'] = 'Introvert Ivan'
    player['energie'] = 70
    player['nalada'] = 70
    player['opilost'] = 0
    player['pricitnost'] = 80
    player['hlad'] = 40
elif volba_postavy == '4':
    player['jmeno'] = 'Gurmet Gábina'
    player['energie'] = 60
    player['nalada'] = 80
    player['opilost'] = 0
    player['pricitnost'] = 75
    player['hlad'] = 90
else:
    print("Neplatná volba postavy. Zvolen Pracoholik Petr jako výchozí.")
    player['jmeno'] = 'Pracoholik Petr'
    player['energie'] = 40
    player['nalada'] = 60
    player['opilost'] = 0
    player['pricitnost'] = 90
    player['hlad'] = 50

print(f"Vybrali jste si postavu: {player['jmeno']}!")
display_stats(player)
````
